```python
import numpy as np
import pandas as pd

df = pd.read_csv('../football.csv')

A = max(df['Composure'])
B = max(df['Reactions'])
M=[]
c = map(lambda x,y: x>=0.9*A and y>=0.9*B, df['Composure'],df['Reactions'])
c=list(c)
for i in range(len(c)):
    if c[i] == True:
        M.append(df['Age'][i])

print(min(M))
```
