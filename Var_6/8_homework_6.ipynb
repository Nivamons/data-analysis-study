```python
def res(df):
    c = len(df.loc[df['release_date'].str[:1] == '7'])
    a = len(df.loc[df['release_date'].str[:1] == '6'])
    b = len(df.loc[df['release_date'].str[:1] == '8'])
    result = a+b+c
    return result

res(df)
```
